```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rating = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
rating = pd.merge(movies,rating).drop(['genres','timestamp'],axis=1)
rating.head()
```

```python
user_rating = rating.pivot_table(index=['userId'],columns=['title'],values='rating')
user_rating.head()
```

remove the movies which has less than 10 user ratings and replace NaN with 0

```python
user_rating = user_rating.dropna(thresh=10,axis=1).fillna(0)
user_rating.head()
```

```python
#similarity matrix
item_similarity_df = user_rating.corr(method='pearson')
item_similarity_df.head(10)
```

```python
def get_similar_movies(movie_name, user_rating):
    similar_score = item_similarity_df[movie_name] * (user_rating-2.5)
    similar_score = similar_score.sort_values(ascending=False)
    return similar_score

```

```python


action_lover = [("10,000 BC (2008)", 5), ("101 Dalmatians (One Hundred and One Dalmatians) (1961)", 3), ("127 Hours (2010)", 4)]
similar_movies = pd.DataFrame()

for movie, rating in action_lover:
    # Get the similar movies as a Series
    similar_scores_series = get_similar_movies(movie, rating)
    # Convert the Series to a DataFrame with a column name
    similar_scores_df = similar_scores_series.to_frame(name=f'{movie}_{rating}')
    # Use pd.concat to append the new DataFrame to similar_movies
    similar_movies = pd.concat([similar_movies, similar_scores_df], axis=1) # Concatenate along columns

similar_movies=similar_movies.T
similar_movies.head()
similar_movies.sum().sort_values(ascending=False)
```
